```python
import matplotlib.pyplot as plt
import pandas as pd

news = pd.read_csv("https://raw.githubusercontent.com/LavanyaPolamarasetty/Datasets/master/Classification/Social_Network_Ads.csv")
news
```

```python
news.isnull().sum()
```

```python
#Line plot
plt.plot(news['Age'], label = 'Age')
plt.plot(news['EstimatedSalary'], label = 'EstimatedSalary')
plt.legend()
plt.show()
```

```python
#Scatter plot
plt.scatter(news["Age"], news["EstimatedSalary"])
plt.xlabel("Age")
plt.ylabel("EstimatedSalary")
plt.show()
```

```python
#Histogram
plt.hist(news["Age"], color = "m")
plt.xlabel("Age")
plt.show()
```

```python
#Bargraph
x = news["Age"]
y = news["EstimatedSalary"]
plt.bar(x,y)
plt.show()
```

```python
#Piechart
plt.pie(news["Age"][:5], labels = news["EstimatedSalary"][:5], autopct = "% 1.2f %%")
plt.show()
```

```python
#Boxplot
plt.boxplot(news["Age"])
plt.show()
```

```python
#Subplot
fig, area = plt.subplots(2, 2)
area[0][0].plot(news['Age'])
area[0][1].plot(news['EstimatedSalary'])
area[1][0].plot(news['Purchased'])
area[1][1].plot(news['Gender'])
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(news['Age'])
plt.subplot(1, 2, 2)
plt.plot(news['EstimatedSalary'])
plt.show()
```
