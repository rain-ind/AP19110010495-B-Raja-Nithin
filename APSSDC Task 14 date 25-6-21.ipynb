## Task 1
- Replace null values 'Age' column with ffill().
- Replace null values in 'Cabin and 'Embarked' columns with most occured value

```python
import pandas as pa
import numpy as nu
```

```python
ship = pa.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/titanic.csv')
ship
```

```python
#Replace null values 'Age' column with ffill().
ship['Age'].fillna(method = 'ffill', inplace = True)
ship
```

### Replace null values in 'Cabin and 'Embarked' columns with most occured value

```python
ship['Cabin'].value_counts()
```

```python
ship['Embarked'].value_counts()
```

```python
ship["Cabin"].fillna("G6", inplace = True)
ship
```

```python
ship["Embarked"].fillna("S", inplace = True)
ship
```

## Task2:
- After handling missing values, Convert Cabin, Embarked columns into numerical columns

```python
from sklearn.preprocessing import LabelEncoder
```

```python
la = LabelEncoder()
```

```python
ship['Embarked'] = la.fit_transform(ship['Embarked'])
ship
```

```python
ship['Cabin'] = la.fit_transform(ship['Cabin'])
ship
```
